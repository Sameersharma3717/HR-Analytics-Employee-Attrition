# 📈 Exploratory Data Analysis (EDA)

This section explores employee demographics, salary distribution, workforce characteristics, and attrition patterns through data visualization.

The objective is to uncover meaningful insights that can support HR decision-making and employee retention strategies.

## 📊 Employee Attrition Distribution

### Business Question

**How many employees have left the organization, and what proportion of the workforce remains?**

Understanding the overall attrition level helps HR evaluate workforce stability and determine whether employee turnover requires immediate attention.

```python
attrition = df["Attrition"].value_counts()

plt.bar(attrition.index, attrition.values)
plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")

plt.show()
```

### 🔍 Observation

- The majority of employees remain with the organization.
- A smaller but significant number of employees have left the company.

### 💡 Business Insight

Although most employees continue to work for the organization, the attrition level indicates opportunities to improve employee retention and reduce turnover-related costs.

### ✅ Recommendation

HR should further investigate the reasons behind employee attrition by analyzing factors such as overtime, job satisfaction, salary, and work-life balance.

## 📊 Department-wise Attrition Analysis

### Business Question

**Which departments experience the highest employee attrition?**

Identifying departments with higher employee turnover enables HR to prioritize retention initiatives and allocate resources effectively.

```python
department_attrition = (
    df[df["Attrition"]=="Yes"]
    .groupby("Department")
    .size()
    .sort_values(ascending=False)
)

plt.bar(department_attrition.index, department_attrition.values)

plt.title("Department-wise Attrition")
plt.xlabel("Department")
plt.ylabel("Employees Left")

plt.show()
```

### 🔍 Observation

- The Research & Development department recorded the highest number of employees leaving the organization.
- The Sales department experienced the second-highest level of attrition.
- The Human Resources department reported the lowest number of employee exits among all departments.

### 💡 Business Insight

Employee turnover is concentrated primarily in the Research & Development and Sales departments. Since these departments represent core business functions, high attrition may affect productivity, project continuity, customer relationships, and increase recruitment and training costs.

### ✅ Business Recommendation

HR should prioritize retention initiatives in the Research & Development and Sales departments by conducting exit interviews, reviewing workload distribution, strengthening career development opportunities, and enhancing employee engagement programs. Continuous monitoring of department-wise attrition can help identify emerging workforce challenges and support proactive decision-making.

## 📊 Gender Distribution

### Business Question

**What is the gender composition of the organization's workforce?**

Understanding workforce diversity supports recruitment planning and diversity & inclusion initiatives.

```python
gender = df["Gender"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(
    gender.values,
    labels=gender.index,
      autopct="%1.1f%%"
)

plt.title("Gender Distribution")

plt.show()
```

### 🔍 Observation

- The organization has a higher proportion of male employees than female employees.
- Male employees represent the majority of the workforce, while female employees account for a smaller share.

### 💡 Business Insight

The workforce demonstrates a gender imbalance, with male employees forming the larger segment of the organization. Monitoring workforce diversity is important for promoting an inclusive workplace and ensuring equal opportunities across all employee groups.

### ✅ Business Recommendation

HR should continue strengthening diversity and inclusion initiatives by promoting equitable recruitment, career development, and leadership opportunities. Regularly tracking gender distribution can help evaluate the effectiveness of diversity policies and support long-term workforce planning.

## 📊 Monthly Income Distribution

Analyzing salary distribution helps identify income concentration and potential compensation gaps.

```python
plt.hist(df["MonthlyIncome"], bins=20)

plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Number of Employees")

plt.show()
```

- The salary distribution is positively skewed, with a small group of employees earning significantly higher incomes.

The organization's workforce is primarily composed of employees in lower and mid-level salary bands, while higher salary levels are limited to a smaller group of senior or specialized roles. This reflects a typical organizational compensation structure with a broader base of operational employees and fewer high-paying positions.

### ✅ Business Recommendation

HR should periodically review salary structures and benchmark compensation against industry standards to ensure competitive pay across all job levels. Clear career progression and transparent compensation policies can help improve employee satisfaction and reduce the risk of attrition among high-performing employees.

## 📊 Job Role Distribution

### Business Question

**How are employees distributed across different job roles within the organization?**

Understanding the workforce composition across job roles helps HR identify critical positions, support workforce planning, and allocate recruitment and training resources effectively.

```python
job_role = df["JobRole"].value_counts()

plt.bar(job_role.index, job_role.values)

plt.title("Employee Distribution by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45,ha="right")
plt.show()
```

- Sales Executives represent the largest employee group in the organization, followed by Research Scientists and Laboratory Technicians.
- Managerial and Research Director positions have comparatively fewer employees, while the Human Resources role has the smallest workforce.
- The workforce distribution indicates that operational and customer-facing roles make up the majority of employees.

### 💡 Business Insight

The organization relies heavily on operational and technical roles to support its core business functions. A larger workforce in these positions means that turnover within these job roles could have a significant impact on productivity, customer service, and overall business performance.

### ✅ Business Recommendation

HR should prioritize employee engagement, career development, and retention strategies for high-volume job roles such as Sales Executives, Research Scientists, and Laboratory Technicians. Workforce planning should also ensure an adequate pipeline of talent for specialized and leadership positions to support long-term organizational growth.

## 📊 Overtime vs Employee Attrition

### Business Question

**Does working overtime influence employee attrition?**

This analysis examines whether employees who work overtime are more likely to leave the organization compared to those who do not.

```python
overtime_attrition = (df[df["Attrition"]=="Yes"]
                    .groupby("OverTime")
                    .size())

plt.bar(overtime_attrition.index,overtime_attrition .values)
plt.title("Overtime vs Employee Attrition")
plt.xlabel("Overtime")
plt.ylabel("Employees Left")

plt.show()
```

### 🔍 Observation

- Employees who worked overtime recorded a higher number of resignations compared to employees who did not work overtime.
- The difference suggests that overtime is associated with increased employee turnover.
- Employees not working overtime experienced comparatively lower attrition.

### 💡 Business Insight

Overtime appears to be an important factor influencing employee attrition. Excessive workloads and longer working hours may contribute to employee stress, reduced work-life balance, and lower job satisfaction, ultimately increasing the likelihood of employees leaving the organization.

### ✅ Business Recommendation

HR and department managers should regularly monitor overtime patterns and ensure workloads are distributed effectively. Promoting work-life balance, optimizing staffing levels, and providing employee wellness initiatives can help reduce overtime-related burnout and improve employee retention.

## 📊 Average Monthly Income by Department

### Business Question

**How does the average monthly income vary across departments?**

Comparing department-wise salaries helps understand compensation patterns and supports strategic workforce planning.

```python
department_salary=(df
            .groupby("Department")["MonthlyIncome"]
            .mean()
            .sort_values(ascending=False))

plt.bar(department_salary.index, department_salary.values)

plt.title("Average Monthly Income by Department")
plt.xlabel("Department")
plt.ylabel("Average Monthly Income")

plt.xticks(rotation=15)

plt.show()
```

### 🔍 Observation

- The Sales department has the highest average monthly income among all departments.
- Human Resources ranks second in terms of average monthly income.
- Research & Development has the lowest average monthly income, although the difference across departments is relatively small.

### 💡 Business Insight

The variation in average monthly income across departments suggests differences in job responsibilities, skill requirements, and compensation structures. Despite having the largest workforce, the Research & Development department has the lowest average salary, which may influence employee satisfaction and retention if compensation is not aligned with market standards.

### ✅ Business Recommendation

HR should periodically review department-wise compensation to ensure salaries remain competitive and equitable. Special attention should be given to departments with comparatively lower average salaries, while considering factors such as job complexity, experience requirements, and industry benchmarks before making compensation decisions.

## 📊 Correlation Matrix

### Business Question

**What relationships exist among the numerical variables in the dataset?**

A correlation matrix helps identify variables that move together and highlights potential factors associated with employee behavior.

```python
correlation = df.select_dtypes(include="number").corr()

plt.figure(figsize=(10,8))

plt.imshow(correlation, cmap="viridis")

plt.colorbar()

plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)

plt.title("Correlation Matrix")

plt.show()
```

### 🔍 Observation

- Most numerical variables show weak to moderate correlations with one another, indicating that they measure different aspects of employee characteristics.
- Strong positive correlations are observed among experience-related variables such as **Total Working Years**, **Years at Company**, **Years in Current Role**, **Years Since Last Promotion**, and **Years with Current Manager**.
- **Monthly Income** shows a positive correlation with **Job Level** and **Total Working Years**, suggesting that employees with higher job levels and greater experience tend to earn higher salaries.

### 💡 Business Insight

The correlation matrix indicates that employee experience and job level are closely associated with salary progression. Additionally, the strong relationships among tenure-related variables reflect expected career progression within the organization. Since most variables have low correlations, employee attrition is likely influenced by multiple factors rather than a single characteristic.

### ✅ Business Recommendation

HR should develop structured career progression and promotion pathways to retain experienced employees and encourage long-term growth within the organization. Further analysis using predictive models or feature importance techniques can help identify the most influential factors contributing to employee attrition and support more targeted retention strategies.
